```python
def get_google_address_ins(x):
    temp = gmaps.find_place(input=x.restaurant_name +' ' +x.address +' '+ x.city_name,input_type = 'textquery',fields = ('formatted_address','permanently_closed','name'))
    if (temp['candidates'] ==''):
        return (temp['candidates'][0]['formatted_address'])
    else:
        return None 
    
```

```python
def get_google_address_ins_test(x):
    temp = gmaps.find_place(input=x.restaurant_name + x.address + x.city_name,input_type = 'textquery',fields = ('formatted_address','permanently_closed','name'))
    return temp
```

```python
def get_google_name_ins(x):
    temp = gmaps.find_place(input=x.restaurant_name + x.address + x.city_name,input_type = 'textquery',fields = ('formatted_address','permanently_closed','name'))
    if (temp['candidates'] ==''):
        return (temp['candidates'][0]['name'])
    else :
        return None
    
```
