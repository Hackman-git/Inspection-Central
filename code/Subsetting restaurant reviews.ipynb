```python
# Because yelp data is hard to work with because of the size let's analyze categories from business data instead
business = pd.read_csv(r"C:/Users/abdul/Desktop/business.csv")
```

```python
business.head()
```

```python
# this is a list of lists for each observation's categpries column
all_cats = []
```

```python
a = list(business.categories)
try:
    for j in range(len(a)):
        all_cats.append(a[j].split(','))
except:
    pass
```

```python
# all possible categories
cats = [j for i in all_cats for j in i]

# strip whitespace
for i in range(len(cats)):
    cats[i] = cats[i].strip()
    
# all unique categories
unique_cats = list(set(cats))
unique_cats
```

```python
# number of unique categories
num_categories = len(unique_cats)
num_categories
```

```python
# categories related to food
food_related = ['Tacos','Custom Cakes', 'Wine & Spirits', 'Sandwiches', 'Hookah Bars', 'Ice Cream & Frozen Yogurt',
                             'Middle Eastern', 'Gastropubs', 'Chicken Wings','Fast Food','Soup','French',
                             'Tex-Mex','Specialty Food','Persian/Iranian','Wine Bars', 'Hot Dogs','Dive Bars', 'Chinese',
                            'Buffets','Barbeque','Beer Gardens', 'Desserts','Cafes','Vietnamese','Burgers','Fish & Chips',
                             'Southern','Pubs', 'Japanese','Vegan','Thai','Imported Food','Pizza','Canadian (New)','Turkish',
                            'Mexican', 'American (Traditional)','Modern European','Bakeries','Bars','Dim Sum','Ethnic Food',
                            'Seafood', 'Delis','Cajun/Creole','Cupcakes','Vegetarian','Diners','Steakhouses','Irish Pub',
                             'Italian','Tapas Bars','American (New)', 'Food Banks','Breakfast & Brunch','Restaurants', 'Donuts',
                             'Juice Bars & Smoothies','Beer','Caterers','Asian Fusion','Salad','Herbs & Spices','Korean',
                             'Food','Mediterranean','Caribbean','Portuguese','Sushi Bars','Coffee & Tea','Comfort Food']
```

```python
# potentially problematic categories
problem = []
for i in range(len(unique_cats)):
    if unique_cats[i] not in food_related:
        problem.append(unique_cats[i])
problem
```

```python
# how many problematic categories are there?
len(problem)
```

```python
# which rows are problematic?
yelp_issues = yelp[yelp.categories.str.contains('|'.join(problem))]
```

```python
# Wnat's the size?
yelp_issues.shape
```

```python
yelp_issues = yelp_issues.reset_index(drop=True)
```

```python
yelp_issues.head(50)
```

```python
yelp_issues.text[12]
```

```python
# Let's remove these potentially problematic records
yelp = yelp[~yelp.review_id.isin(yelp_issues.review_id)]
yelp.shape
```
