### Join the two datasets

Take the mapping data set and merge it with the establishments data.

Take that joined data and merge with the business data.

Should end up with a dataframe where each row is a business with an inspection. There are going to be multiple business ids with multiple rows. That means the business has been inspected multiple times

```python
# Import the mapping file and only use the valid matches
# Remove some columns and rename another column for joining purposes
mapping_df = pd.read_csv(file_path + "data/mapping_file.csv")
mapping_df = mapping_df.drop(columns='Unnamed: 8')
mapping_df = mapping_df[mapping_df['is_valid'] == 'y']
mapping_df = mapping_df.reset_index(drop = True)
mapping_df = mapping_df.rename(columns={'address_y': 'address'})
mapping_df.shape
```

```python
mapping_df.head()
```

```python
# We inner join because we only want establishments that have a matching business on the yelp dataset
# and we only care about the mappings that actually map to an establishment
establishment_with_map_df = establishment_df.merge(mapping_df, how='inner', 
                                                   on=['facility_id', 'location_name', 'restaurant_name', 'address'])
```

```python
establishment_with_map_df.shape
```

```python
establishment_with_map_df.head()
```

```python
# We want an inner join because we only care about businesses that have inspection data
business_ment_df = business_df.merge(establishment_with_map_df, how = "inner", 
                                     on='business_id',
                                     suffixes=('_left', '_right'))
```

```python
business_ment_df.isna().sum()
```

```python
business_ment_df = business_ment_df.dropna(axis='index', subset=['current_grade'])
business_ment_df = business_ment_df.drop(columns=['nciaa'])
business_ment_df.isna().sum()
```

```python
from datetime import datetime, timedelta

time_diff = timedelta(days=30)

unique_inspections = business_ment_df.groupby(['facility_id']).agg({'date_current': pd.Series.nunique})
facility_multiple_inspections = unique_inspections[unique_inspections['date_current'] > 1].index
```

```python
#TODO: just apply the strptime function to the whole column

for facility in facility_multiple_inspections:
    facility_rows = business_ment_df[business_ment_df['facility_id'] == facility]
    inspection_dates = [datetime.strptime(x, '%Y-%m-%d %H:%M:%S') for x in 
                        sorted(facility_rows['date_current'].drop_duplicates().tolist())]

    for idx, date in enumerate(inspection_dates):
        if idx == len(inspection_dates) - 1:
            print("idx limit", False)
        else:
            if (date + time_diff) > inspection_dates[idx + 1]:
                print(date + time_diff, inspection_dates[idx + 1], True)
            else:
                print(False)
```
