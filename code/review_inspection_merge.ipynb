### Join the two datasets

Take the mapping data set and merge it with the establishments data.

Take that joined data and merge with the business data.

Should end up with a dataframe where each row is a business with an inspection. There are going to be multiple business ids with multiple rows. That means the business has been inspected multiple times

```python
# Import the mapping file and only use the valid matches
# Remove some columns and rename another column for joining purposes
mapping_df = pd.read_csv(file_path + "data/mapping_file.csv")
mapping_df = mapping_df.drop(columns='Unnamed: 8')
mapping_df = mapping_df[mapping_df['is_valid'] == 'y']
mapping_df = mapping_df.reset_index(drop = True)
mapping_df = mapping_df.rename(columns={'address_y': 'address'})
mapping_df.shape
```

```python
mapping_df.head()
```
