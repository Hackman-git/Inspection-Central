```python
# Taking out the 163 instances of NaNs for categories
yelp = yelp[yelp.categories.notna()]
```

```python
yelp.shape
```

```python
'''
We can take advantage of the categories attribute to extract only columns with 
"Restaurants" as one of the categories (Yelp did a good job of properly tagging reviews)
'''
x = yelp[yelp.categories.str.contains('Restaurant' or 'food' or 'bar')]
```

```python
x.shape
```

```python
x.head()
```

```python
yelp = x
```

```python
import nltk
```

```python
nltk.download()
```

```python
yelp['tokenized'] = yelp.apply(lambda obs: nltk.word_tokenize(obs.text), axis=1)
```

```python
yelp.head()
```

```python
yelp.iloc[1].tokenized
```

```python
yelp = yelp.reset_index()
```

```python
yelp['review_length'] = yelp.apply(lambda obs: len(obs.tokenized), axis=1)
```

```python
yelp.head()
```

```python
# histogram of word length
import seaborn as sns
import matplotlib.pyplot as plt


plt.figure(figsize=(20,8))
sns.distplot(yelp.review_length, kde=False, color='blue')
plt.show()
```
