### Combining necessary datasets in the Inspection Data

```python
Inspections = pd.read_csv(r"C:/Users/abdul/Desktop/restaurant_inspections.csv", delimiter=";")
```

```python
Inspections.head(10)
```

```python
# Cleaning
Inspections.drop(columns=["Unnamed: 15"], axis=1, inplace=True)
```

```python
# Because this is the main data, we have to maintain this number of rows
Inspections.shape
```

```python
# Reading inspection type data
Inspection_types = pd.read_csv(r"C:/Users/abdul/Desktop/restaurant_inspection_types.csv", delimiter=";")
```

```python
Inspection_types.head()
```

```python
# cleaning
Inspection_types.drop(Inspection_types.columns[2], axis=1, inplace=True)
Inspection_types.head()
```

```python
# performing a join to get inspection types
Inspections = pd.merge(Inspections, Inspection_types, on="inspection_type_id")
```

```python
Inspections.head()
```

```python
# confirming shape
Inspections.shape
```

```python
# Getting establishment info
Establishments = pd.read_csv(r"C:/Users/abdul/Desktop/restaurant_establishments.csv", delimiter=";")
```

```python
Establishments.head()
```

```python
# cleaning
Establishments.drop(columns=["Unnamed: 21"], inplace=True)
```

```python
Establishments.shape
```

```python
Establishments.head()
```

```python
# adding establishment info to the main data(Inspections)
Inspections = pd.merge(Inspections, Establishments[["permit_number", "facility_id", "PE", "restaurant_name", "location_name",
                                                   "address", "longitude", "latitude", "city_id", "city_name", 
                                                    "zip_code", "nciaa"]], on="permit_number", how="left")
```
