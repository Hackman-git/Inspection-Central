```python
yelp[yelp.categories.isna()]
```

```python
# check
yelp.categories.isna().sum()
```

```python
# Hours column NAs
yelp.hours.isna().sum()
```

```python
# That's quite a lot but hours of operation isn't going to add a lot
# of predictive power either so I could care less. The median is most likely 9am-6pm thereabout.
# I'll replace with 'none' for now
yelp.hours.fillna('none', inplace=True)
```

```python
# check
yelp.hours.isna().sum()
```
