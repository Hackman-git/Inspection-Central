### Tokenization

```python
import pandas as pd
```

```python
yelp = pd.read_csv(r"C:/Users/abdul/Desktop/yelp_stage2.csv")
inspections = pd.read_csv(r"C:/Users/abdul/Desktop/Inspections_stage2.csv")
```

```python
import nltk
```

```python
# nltk.download()
```

```python
yelp['tokenized'] = yelp.apply(lambda obs: nltk.word_tokenize(obs.text), axis=1)
```

```python
yelp.head()
```

```python
yelp.iloc[1].tokenized
```

```python
yelp = yelp.reset_index()
```

```python
yelp['review_length'] = yelp.apply(lambda obs: len(obs.tokenized), axis=1)
```

```python
yelp.head()
```

```python
# histogram of word length
import seaborn as sns
import matplotlib.pyplot as plt


plt.figure(figsize=(20,8))
sns.distplot(yelp.review_length, kde=False, color='blue')
plt.show()
```

```python
yelp.review_length.mean()
```

```python
yelp.review_length.median()
```
