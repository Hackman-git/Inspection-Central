```
def clean_text(x,punc =True,stop =True):
    temp = str(x)
    temp = temp.lower()
    tokens = word_tokenize(temp)
    tokens= list(map(lambda b: b.replace("n't","not"), tokens))
    words =tokens
    if (punc==True):
        words = [word for word in tokens if word.isalpha()]
    if (stop == True) :   
        words = [w for w in words if not w in stop_words]    

    return words 
```

```
def get_lda_corpus(list_text):
    data_processed = []
    i = 0
    for doc in list_text :
        doc_out = []
        for wd in clean_text(doc):
            if wd not in stop_words:  # remove stopwords
                lemmatized_word = lemmatize(wd, allowed_tags=re.compile('(NN|JJ|RB)'))  # lemmatize
                if lemmatized_word:
                    doc_out = doc_out  + [(lemmatized_word[0].split(b'/')[0].decode('utf-8'))]
                else:
                    continue
        data_processed.append(doc_out)
    return data_processed  
```

```
def get_lda_corpus_text(doc):
    list_clean  = clean_text(doc)
    doc_out = []
    for wd in list_clean:
        if (wd != 'not'):  # remove stopwords
            lemmatized_word = lemmatize(wd, allowed_tags=re.compile('(NN|JJ|RB)'))  # lemmatize
            if lemmatized_word:
                  doc_out = doc_out + ([lemmatized_word[0].split(b'/')[0].decode('utf-8')])
            else:
                continue
        else:
          doc_out = doc_out + [wd]
    i = 0    
    for i in range(len(doc_out)-2):
        if doc_out[i] == 'not' or doc_out[i] == "n't":
           doc_out[i+1] = 'neg_' + doc_out[i+1] 
           doc_out[i+2] = 'neg_' + doc_out[i+2]
    while 'not' in doc_out:
        doc_out.remove('not')                          
    return(doc_out)  
```

```
review_dataset['text_list'] = review_dataset.text.parallel_apply(lambda x:get_lda_corpus_text(x))
```

```
dct = corpora.Dictionary(review_dataset.text_list)
corpus = [dct.doc2bow(doc) for doc in review_dataset.text_list]
```
